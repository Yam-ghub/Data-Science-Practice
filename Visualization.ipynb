```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Ice Cream Ratings.csv")
df = df.set_index("Date")
df
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating',  ylabel = 'Scores')
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.scatter( x = 'Texture Rating', y = 'Overall Rating', s = 200, c = 'violet')
```

```python
df.plot.hist (bins = 20)
```
